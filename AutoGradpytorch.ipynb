```python
x =torch.tensor(6.7)
y =torch.tensor(0.0)

w =torch.tensor(1.0)
b=torch.tensor(0.0)
```

```python
#Binary cross entropy
def binary_cross_entropy(prediction,target):
  epsilon = 1e-8
  prediction = torch.clamp(prediction,epsilon,1-epsilon)
  loss = -(target*torch.log(prediction)+(1-target)*torch.log(1-prediction))
  return loss
```

```python
#Forward Pass
z =w*x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy(y_pred,y)
```

```python
loss
```

```python
#Backpropogation
#dL/d(y_pred) =Loss w.r.t prediction
dloss_y_pred =(y_pred -y)/(y_pred-(1-y_pred))

#dy_pred/dz =prediction(y_pred) with respect to Z
dy_pred_dz = y_pred*(1-y_pred)

#dz/dw =x
dz_dw =x

#dz/db =1
dz_db =1

dL_dw =dloss_y_pred*dy_pred_dz*dz_dw
dL_db =dloss_y_pred*dy_pred_dz*dz_db
```

```python
print(f"mannual gradient of loss w.r.t weight(dw):{dL_dw}")
print(f"mannual gradient of loss w.r.t bias(db):{dL_db}")
```

#using autograd

```python
x =torch.tensor(6.7,requires_grad=True)
y =torch.tensor(0.0)

w =torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
```

```python
z = w*x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy(y_pred, y)

loss.backward()
```

```python
print(w.grad)
print(b.grad)
```
